finding numerical columns

```python
num_col=df.select_dtypes(include=[np.number])
num_col
```

```python
mean=[]
median=[]
mode=[]
std=[]
for i in num_col:
   mean.append(num_col[i].mean())
   median.append(num_col[i].median())
   mode.append(num_col[i].mode())
   std.append(num_col[i].std())
   #print(num_col[i])
```

```python
print("Mean=",mean)
print("Median=",median)
print("Standard deviation=",std)
print("Mode=",mode)
```

Histograms

```python
figb, axesb = plt.subplots(3, 2, figsize=(16, 10))
# Plotting the histograms 
(df['Discount Amount']).plot(kind='hist', ax=axesb[0,0])
axesb[0,0].set_title('Discount Amount')

(df['Avg Price']).plot(kind='hist', ax=axesb[0,1])
axesb[0,1].set_title('Avg Price')

(df['Volume']).plot(kind='hist', ax=axesb[1,0])
axesb[1,0].set_title('Volume')

(df['Net Sales Value']).plot(kind='hist', ax=axesb[1,1])
axesb[1,1].set_title('Net Sales Value')

(df['Total Sales Value']).plot(kind='hist', ax=axesb[2,0])
axesb[2,0].set_title('Total Sales Value')

(df['Net Sales Value']).plot(kind='hist', ax=axesb[2,1])
axesb[2,1].set_title('Net Sales Value')
figb.tight_layout()
plt.show()
```

Box plots

```python
figd, axesd = plt.subplots(3, 2, figsize=(16, 10))
# Plotting the histograms 
(df['Discount Amount']).plot(kind='box', ax=axesd[0,0])
axesd[0,0].set_title('Discount Amount')

(df['Avg Price']).plot(kind='box', ax=axesd[0,1])
axesd[0,1].set_title('Avg Price')

(df['Volume']).plot(kind='box', ax=axesd[1,0])
axesd[1,0].set_title('Volume')

(df['Net Sales Value']).plot(kind='box', ax=axesd[1,1])
axesd[1,1].set_title('Net Sales Value')

(df['Total Sales Value']).plot(kind='box', ax=axesd[2,0])
axesd[2,0].set_title('Total Sales Value')

(df['Net Sales Value']).plot(kind='box', ax=axesd[2,1])
axesd[2,1].set_xlim([ df['Net Sales Value']])
figd.tight_layout()
plt.show()
```
